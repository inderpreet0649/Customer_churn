```python
import matplotlib.pyplot as plt
import pandas as pd

# Step 1: Create a crosstab (counts)
ct = pd.crosstab(df['SeniorCitizen'], df['Churn'])

# Step 2: Convert counts to row-wise percentages
ct_percent = ct.div(ct.sum(axis=1), axis=0) * 100

# Step 3: Plot
ax = ct_percent.plot(kind='bar', stacked=True, figsize=(4, 4), color=['skyblue', 'salmon'])

# Step 4: Add percentage labels
for i, row in enumerate(ct_percent.values):
    cumulative = 0
    for j, val in enumerate(row):
        if val > 0:
            ax.text(
                i,
                cumulative + val / 2,
                f'{val:.1f}%',
                ha='center',
                va='center',
                fontsize=8,
                color='black'
            )
            cumulative += val

# Step 5: Customize plot
plt.title('Customer Churn Rate by Senior Citizen Status', fontsize=10, fontname='Bell MT')  # Optional: custom font
plt.xlabel('Senior Citizen')
plt.ylabel('Percentage')
plt.xticks(ticks=[0, 1], labels=['No', 'Yes'], rotation=0)
plt.ylim(0, 100)
plt.legend(title='Churn')
plt.tight_layout()
plt.show()
```

```python
n_col = 3
n_rows = 5

fig, axes = plt.subplots(n_rows, n_col, figsize=(16, 16))
axes = axes.flatten()

for i in range(len(columns)):
    ax = sns.countplot(data=df, x=columns[i], hue='Churn', ax=axes[i])
    axes[i].set_title(f'Count Plot of {columns[i]}')

    # Get total count per group for percentage calculation
    total_per_category = df[columns[i]].value_counts()

    for container in ax.containers:
        # Compute total count for each bar group (hue value)
        labels = []
        for bar in container:
            height = bar.get_height()
            x_val = bar.get_x() + bar.get_width() / 2
            x_category = round(x_val)  # approximate index for x-axis category
            
            # Get category name from bar's x position
            category_idx = int(bar.get_x() + bar.get_width() / 2)
            category_name = bar.get_x()
            
            # Get corresponding x-tick label (safer way)
            xtick_labels = [tick.get_text() for tick in ax.get_xticklabels()]
            if xtick_labels:
                category_name = xtick_labels[container.index(bar) % len(xtick_labels)]

            # Compute percentage
            total = df[df[columns[i]] == category_name].shape[0]
            percent = (height / total) * 100 if total > 0 else 0
            labels.append(f'{percent:.1f}%')

        ax.bar_label(container, labels=labels)
plt.tight_layout()
```

# Key Insights 

#### SeniorCitizen
- 23.6% of senior citizens churned, higher than the 16.4% of non-seniors, indicating potential dissatisfaction among older customers.

#### Partner
- 19.7% of customers with partners churned, compared to a lower rate among those without, suggesting relationship status may influence retention.

#### Dependents
- 15.5% of customers with dependents churned, less than those without (31.3%), indicating dependents might stabilize customer loyalty.

#### PhoneService
- 26.7% of customers without phone service churned, slightly more than the 24.9% with it, pointing to limited service uptake as a factor.

#### MultipleLines
- 28.6% of customers with multiple lines churned, higher than the 24.9% without, suggesting complexity may lead to dissatisfaction.

#### InternetService
- Fiber optic users show a 19.0% churn rate, significantly higher than DSLâ€™s 7.4%, highlighting potential service quality issues.

#### OnlineSecurity
- 14.6% churn without online security versus 7.4% with it, indicating security concerns drive customer departures.

#### OnlineBackup
- 7.4% churn without online backup (no internet service) versus higher rates with it, suggesting backup may not be a key retention factor.

#### DeviceProtection
- 7.4% churn without device protection (no internet service) versus higher rates with it, indicating limited impact on retention.

#### TechSupport
- 15.2% churn without tech support versus 7.4% with it, showing the importance of support services.

#### StreamingTV
- 7.4% churn without streaming TV (no internet service) versus higher rates with it, suggesting limited retention value.

#### StreamingMovies
- 7.4% churn without streaming movies (no internet service) versus higher rates with it, indicating minimal impact on churn.

#### Contract
- 42.7% churn with month-to-month contracts compared to 2.8% with two-year contracts, emphasizing the role of contract duration.

#### PaperlessBilling
- 16.3% churn without paperless billing versus 33.6% with it, suggesting billing preferences affect retention.

#### PaymentMethod
- 19.1% churn with electronic checks versus 15.2% with bank transfers, indicating payment method influences churn rates.

# Conclusion
Customer churn is significantly influenced by contract duration, with month-to-month users leaving at the highest rate. The absence of additional services like online security and variations in internet service quality (e.g., fiber optic vs. DSL) also contribute to higher churn. Demographic factors, such as senior citizens, and lack of phone service further exacerbate the issue.

# Suggestions

Offer incentives for longer contracts to reduce month-to-month churn.
Promote additional services like online security and tech support to enhance customer retention.
Improve fiber optic service quality to match DSL satisfaction levels.
Provide tailored support for senior citizens to address their specific needs.
Encourage adoption of phone services to increase customer engagement.
